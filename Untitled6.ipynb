```python
%%writefile app.py
import streamlit as st
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import kagglehub
import arabic_reshaper
from bidi.algorithm import get_display
import plotly.express as px
import base64

st.set_page_config(page_title="Saudi Real Estate EDA", layout="wide")

st.markdown("""
<style>
    .stApp { background-color: #0a1f0a; }
    .stSidebar { background-color: #0d2b0d; }
    .metric-card {
        background-color: #0d2b0d;
        border: 1px solid #008236;
        border-radius: 10px;
        padding: 20px;
        text-align: center;
    }
    .metric-value { font-size: 28px; font-weight: bold; color: #008236; }
    .metric-label { font-size: 14px; color: #C5A028; }
    h1, h2, h3 { color: #ffffff; }
    .stMultiSelect span { background-color: #008236 !important; color: white !important; }
    .stMultiSelect [data-baseweb="tag"] { background-color: #008236 !important; }
</style>
""", unsafe_allow_html=True)

RIYADH_NIGHTS = [
    '#008236', '#C5A028', '#1B3A6B', '#D4B896', '#8B4513',
    '#2E8B57', '#DAA520', '#4169E1', '#F5DEB3', '#A0522D',
    '#006400', '#FFD700'
]

fm.fontManager.addfont('Amiri.ttf')
plt.rcParams['font.family'] = 'Amiri'
plt.rcParams['figure.facecolor'] = '#0a1f0a'
plt.rcParams['axes.facecolor'] = '#0d2b0d'
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = '#C5A028'
plt.rcParams['xtick.color'] = '#C5A028'
plt.rcParams['ytick.color'] = '#C5A028'
plt.rcParams['axes.edgecolor'] = '#008236'
plt.rcParams['grid.color'] = '#1a3a1a'

def fix_arabic(text):
    return get_display(arabic_reshaper.reshape(str(text)))

city_map_en = {
    'الرياض': 'Riyadh', 'جدة': 'Jeddah', 'الدمام': 'Dammam',
    'الخبر': 'Khobar', 'المدينة المنورة': 'Madinah', 'مكة المكرمة': 'Makkah',
    'بريدة': 'Buraidah', 'خميس مشيط': 'Khamis Mushait',
    'الهفوف': 'Hofuf', 'الطائف': 'Taif', 'تبوك': 'Tabuk',
    'أبها': 'Abha', 'القصيم': 'Qassim', 'حائل': 'Hail',
    'الجبيل': 'Jubail', 'الاحساء': 'Al-Ahsa', 'نجران': 'Najran',
    'جازان': 'Jazan', 'ينبع': 'Yanbu', 'سكاكا': 'Sakaka'
}

@st.cache_data
def load_data():
    path = kagglehub.dataset_download("mohdph/saudi-arabia-real-estate-dataset")
    conn = sqlite3.connect(path + "/database.db")
    df = pd.read_sql_query("SELECT * FROM Listings", conn)
    conn.close()
    cols_to_drop = ['native.logo','native.title','native.image','native.description',
                    'native.external_url','has_extended_details','rent_period','ac',
                    'user.rega_id','advertiser_type','daily_rentable','uri','imgs',
                    'content','path','user.img','user.review','width','length',
                    'user.name','user.iam_verified','user.phone','user_id','id',
                    'refresh','last_update','create_time','city_id','district_id']
    df = df.drop(columns=cols_to_drop, errors='ignore')
    df = df[(df['price'] >= 10000) & (df['price'] <= 500000000)]
    df = df[(df['area'] > 0) & (df['area'] < 100000)]
    for col in ['beds','livings','wc','age','street_width']:
        df[col] = df[col].fillna(df[col].median())
    for col in ['type','furnished','ketchen','street_direction']:
        df[col] = df[col].fillna(df[col].mode()[0])
    df = df.dropna(subset=['title'])
    category_map = {1:'Apartment Rent',2:'Land Sale',3:'Villa Sale',
                    4:'Apartment Sale',5:'House Sale',6:'Flat Sale',
                    7:'Building Sale',8:'Floor Sale',9:'Farm Sale',
                    10:'Chalet Sale',22:'Office Rent'}
    df['category_en'] = df['category'].map(category_map).fillna('Other')
    return df

df = load_data()

with open("/content/KDL6N_sL_400x400.jpeg", "rb") as f:
    img_data = base64.b64encode(f.read()).decode()

col1, col2 = st.columns([1, 6])
with col1:
    st.markdown(f'<img src="data:image/jpeg;base64,{img_data}" width="120">', unsafe_allow_html=True)
with col2:
    st.title("Saudi Real Estate EDA")
    st.subheader("Exploratory Data Analysis — Aqar Dataset")

st.markdown("---")

st.sidebar.title("Filters")
all_cities = sorted(df['city'].unique().tolist())
selected_cities = st.sidebar.multiselect("Select Cities", all_cities,
                                          default=['الرياض', 'جدة', 'الدمام'])
if selected_cities:
    df = df[df['city'].isin(selected_cities)]

st.sidebar.markdown("---")
charts = st.sidebar.multiselect("Select Charts",
    ['Price Distribution by City',
     'Price by Property Type',
     'Top Cities by Average Price',
     'Correlation Heatmap',
     'Average Price by City and Property Type'],
    default=['Price Distribution by City',
             'Price by Property Type',
             'Top Cities by Average Price',
             'Correlation Heatmap',
             'Average Price by City and Property Type'])

c1, c2, c3, c4 = st.columns(4)
with c1:
    st.markdown(f'<div class="metric-card"><div class="metric-value">{df.shape[0]:,}</div><div class="metric-label">Total Listings</div></div>', unsafe_allow_html=True)
with c2:
    st.markdown(f'<div class="metric-card"><div class="metric-value">{df["city"].nunique()}</div><div class="metric-label">Cities</div></div>', unsafe_allow_html=True)
with c3:
    avg_price = df['price'].mean()
    st.markdown(f'<div class="metric-card"><div class="metric-value">{avg_price/1e6:.1f}M</div><div class="metric-label">Avg Price (SAR)</div></div>', unsafe_allow_html=True)
with c4:
    max_price = df['price'].max()
    st.markdown(f'<div class="metric-card"><div class="metric-value">{max_price/1e6:.0f}M</div><div class="metric-label">Max Price (SAR)</div></div>', unsafe_allow_html=True)

st.markdown("<br>", unsafe_allow_html=True)

Q3 = df['price'].quantile(0.75)
IQR = Q3 - df['price'].quantile(0.25)
df_plot = df[df['price'] <= Q3 + 1.5 * IQR].copy()
df_plot['city_fixed'] = df_plot['city'].map(city_map_en).fillna(df_plot['city'])

if 'Price Distribution by City' in charts:
    st.subheader("Price Distribution by City")
    fig, ax = plt.subplots(figsize=(12, 5))
    sns.boxplot(data=df_plot, x='city_fixed', y='price',
                palette=RIYADH_NIGHTS, width=0.5, linewidth=1.5,
                flierprops=dict(marker='o', markerfacecolor='white', markersize=3, alpha=0.3),
                ax=ax)
    ax.set_xlabel('City')
    ax.set_ylabel('Price (SAR)')
    for spine in ax.spines.values():
        spine.set_edgecolor('#008236')
    ax.yaxis.grid(True, linewidth=0.8)
    ax.set_axisbelow(True)
    plt.tight_layout()
    st.pyplot(fig)

if 'Price by Property Type' in charts:
    st.subheader("Price by Property Type")
    fig2, ax2 = plt.subplots(figsize=(14, 5))
    sns.boxplot(data=df_plot, x='category_en', y='price',
                palette=RIYADH_NIGHTS, width=0.5, linewidth=1.5,
                flierprops=dict(marker='o', markerfacecolor='white', markersize=3, alpha=0.3),
                ax=ax2)
    ax2.set_xlabel('Property Type')
    ax2.set_ylabel('Price (SAR)')
    plt.xticks(rotation=45)
    for spine in ax2.spines.values():
        spine.set_edgecolor('#008236')
    ax2.yaxis.grid(True, linewidth=0.8)
    ax2.set_axisbelow(True)
    plt.tight_layout()
    st.pyplot(fig2)

if 'Top Cities by Average Price' in charts:
    st.subheader("Top Cities by Average Price")
    city_avg = df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)
    city_avg.index = [city_map_en.get(c, c) for c in city_avg.index]
    fig3, ax3 = plt.subplots(figsize=(14, 5))
    ax3.bar(city_avg.index, city_avg.values,
            color=RIYADH_NIGHTS[:len(city_avg)], edgecolor='#0a1f0a')
    ax3.set_ylabel('Average Price (SAR)')
    plt.xticks(rotation=45, ha='right', fontsize=11)
    ax3.yaxis.grid(True, linewidth=0.8)
    ax3.set_axisbelow(True)
    plt.tight_layout()
    st.pyplot(fig3)

if 'Correlation Heatmap' in charts:
    st.subheader("Correlation Heatmap")
    cols = ['price','area','beds','livings','wc','age','street_width']
    fig4, ax4 = plt.subplots(figsize=(8, 6))
    sns.heatmap(df[cols].corr(), annot=True, fmt='.2f', cmap='Greens',
                ax=ax4, linewidths=0.5, linecolor='#008236',
                annot_kws={'color': 'white', 'size': 10})
    plt.tight_layout()
    st.pyplot(fig4)

if 'Average Price by City and Property Type' in charts:
    st.subheader("Average Price by City and Property Type")
    pivot_df = df.groupby(['city','category_en'])['price'].mean().reset_index()
    pivot_df['city'] = pivot_df['city'].map(city_map_en).fillna(pivot_df['city'])
    fig5 = px.bar(pivot_df, x='city', y='price', color='category_en',
                  barmode='group',
                  color_discrete_sequence=RIYADH_NIGHTS,
                  labels={'price': 'Average Price (SAR)', 'city': 'City'},
                  template='plotly_dark')
    fig5.update_layout(paper_bgcolor='#0a1f0a', plot_bgcolor='#0d2b0d')
    st.plotly_chart(fig5, use_container_width=True)
```
