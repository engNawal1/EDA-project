```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import arabic_reshaper
from bidi.algorithm import get_display
import matplotlib.font_manager as fm


fm.fontManager.addfont('Amiri.ttf')
plt.rcParams['font.family'] = 'Amiri'
plt.rcParams['figure.facecolor'] = '#0d1117'
plt.rcParams['axes.facecolor'] = '#161b22'
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = '#8b949e'
plt.rcParams['xtick.color'] = '#8b949e'
plt.rcParams['ytick.color'] = '#8b949e'
plt.rcParams['axes.edgecolor'] = '#30363d'
plt.rcParams['grid.color'] = '#21262d'

def fix_arabic(text):
    return get_display(arabic_reshaper.reshape(str(text)))


conn = sqlite3.connect(path + "/database.db")
df = pd.read_sql_query("SELECT * FROM Listings", conn)
conn.close()

print("كل الداتا:", df.shape)
print(df['city'].value_counts().head(10))
```

```python
#delete unusful data

cols_to_drop = ['native.logo','native.title','native.image','native.description',
                'native.external_url','has_extended_details','rent_period','ac',
                'user.rega_id','advertiser_type','daily_rentable','uri','imgs',
                'content','path','user.img','user.review','width','length',
                'user.name','user.iam_verified','user.phone','user_id','id',
                'refresh','last_update','create_time','city_id','district_id']
df = df.drop(columns=cols_to_drop, errors='ignore')
df = df[(df['price'] >= 10000) & (df['price'] <= 500000000)]#make sure the price is real
df = df[(df['area'] > 0) & (df['area'] < 100000)]#make sure about the area
for col in ['beds','livings','wc','age','street_width']:
    df[col] = df[col].fillna(df[col].median())
for col in ['type','furnished','ketchen','street_direction']:
    df[col] = df[col].fillna(df[col].mode()[0])
df = df.dropna(subset=['title'])

category_map = {
    1: 'Apartment Rent',
    2: 'Land Sale',
    3: 'Villa Sale',
    4: 'Apartment Sale',
    5: 'House Sale',
    6: 'Flat Sale',
    7: 'Building Sale',
    8: 'Floor Sale',
    9: 'Farm Sale',
    10: 'Chalet Sale',
    22: 'Office Rent'
}
df['category_en'] = df['category'].map(category_map).fillna('Other')
print(df['category_en'].value_counts())


print(" بعد التنظيف:", df.shape)
print(df['city'].value_counts().head(10))
```

```python
# Column Names
print("Step 5: Column Names")
print(df.columns.tolist())

# General Info
print("\nStep 7: General Info")
df.info()

#Summary Stats
print("\nStep 8: Summary Stats")
print(df.describe())

#Unique Values
print("\nStep 11: Unique Values per Column")
print(df.nunique())

#Column Types
print("\nStep 12: Column Types")
num_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
print("Numerical:", num_cols)
print("Categorical:", cat_cols)
```

```python
# Numerical Distribution
num_cols = ['price', 'area', 'beds', 'livings', 'wc', 'age', 'street_width']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

colors = ['#6C5CE7','#00B894','#FD79A8','#FDCB6E','#E17055','#74B9FF','#A29BFE']

for i, col in enumerate(num_cols):
    axes[i].hist(df[col], bins=30, color=colors[i], edgecolor='#0d1117', alpha=0.8)
    axes[i].set_title(col, color='white', fontsize=12)
    axes[i].set_xlabel('Value', color='#8b949e')
    axes[i].set_ylabel('Count', color='#8b949e')

for j in range(len(num_cols), len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Numerical Columns Distribution', color='white', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Categorical Analysis

for col in ['city', 'category_en']:


    top = df[col].value_counts().head(10)


    new_index = []
    for i in top.index:
        new_index.append(fix_arabic(i))
    top.index = new_index

    if col == 'city':
        title = 'Top 10 Cities'
    else:
        title = 'Top 10 Property Types'


    fig, ax = plt.subplots(figsize=(12, 5))
    ax.bar(top.index, top.values, color='#6C5CE7', edgecolor='#0d1117')
    ax.set_title(title, color='white', fontsize=14)
    ax.set_ylabel('Count', color='#8b949e')
    plt.xticks(rotation=45)
    ax.yaxis.grid(True, color='#21262d', linewidth=0.8)
    ax.set_axisbelow(True)
    plt.tight_layout()
    plt.show()
```

```python
# Top 10 cities by Average Price
city_avg = df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)

new_index = []
for city in city_avg.index:
    new_index.append(fix_arabic(city))
city_avg.index = new_index

fig, ax = plt.subplots(figsize=(14, 6))

colors_list = ['#FD79A8'] * 3 + ['#6C5CE7'] * 12

ax.bar(city_avg.index, city_avg.values,
       color=colors_list, edgecolor='#0d1117')

ax.set_title('Top 10 Cities by Average Price', fontsize=16,
             fontweight='bold', color='white', pad=20)
ax.set_ylabel('Average Price (SAR)', fontsize=12, color='#8b949e')
ax.set_xlabel('City', fontsize=12, color='#8b949e')
plt.xticks(rotation=45, ha='right', fontsize=11)
ax.yaxis.grid(True, color='#21262d', linewidth=0.8)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

```python
# Average Price by City and Property Type
pivot = df.groupby(['city', 'category_en'])['price'].mean().unstack()


new_index = []
for city in pivot.index:
    new_index.append(fix_arabic(city))
pivot.index = new_index

fig, ax = plt.subplots(figsize=(18, 7))
pivot.plot(kind='bar', ax=ax, colormap='tab20')

ax.set_title('Average Price by City and Property Type', fontsize=16,
             fontweight='bold', color='white', pad=20)
ax.set_ylabel('Average Price (SAR)', fontsize=12, color='#8b949e')
ax.set_xlabel('City', fontsize=12, color='#8b949e')
ax.tick_params(colors='#8b949e')
plt.xticks(rotation=45, ha='right', fontsize=9)
ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1),
          facecolor='#1e2433', labelcolor='white', edgecolor='#30363d',
          fontsize=9)

for spine in ax.spines.values():
    spine.set_edgecolor('#30363d')

ax.yaxis.grid(True, color='#21262d', linewidth=0.8)
ax.set_axisbelow(True)
plt.tight_layout()
plt.show()
```

# EDA Observation

- Dataset: 773,643 rows | 18 columns
- Target column: price
- No missing values after cleaning
- price is right-skewed (outliers at high end)
- area is the strongest predictor of price (0.58)
- Riyadh has highest average price (~3.3M SAR)
- Jeddah average price (~1.6M SAR)
- Dammam average price (~1.4M SAR)
- Building Sale is the most expensive property type
- Apartment Rent is the cheapest property type
- Villa Sale and Land Sale are the most common listings
